```python
#EXERCISE 1: Unstructured Data Aggregator

from collections import defaultdict, Counter

#Simulated stream of applications
stream = [
    ("TUP Manila", "BSECE"),
    ("TUP Taguig", f"BSME_v{CONTROL_NUM}"),
    ("TUP Manila", f"BSME_v{CONTROL_NUM}"),
    ("TUP Manila", "BSECE"),
    ("TUP Visayas", STUDENT_MAJOR),
    ("TUP Taguig", "BSECE"),
    ("TUP Manila", "BSECE")
]

#Data Expansion: add CONTROL_NUM + 3 = 5 custom applications
stream.extend([
    ("TUP Manila", FAVORITE_ARTIST),
    ("TUP Taguig", LOCAL_HAZARD),
    ("TUP Visayas", "BSCE"),
    ("TUP Manila", "BSME"),
    ("TUP Taguig", "BSIT")
])

#Phase 1: Aggregation
campus_programs = defaultdict(list)
for campus, program in stream:
    campus_programs[campus].append(program)

#Phase 2: Frequency Analysis
manila_counter = Counter(campus_programs["TUP Manila"])
top_program, top_freq = manila_counter.most_common(1)[0]   # unpack tuple properly

#PRINT ASSESSMENT DATA
print("=== ASSESSMENT DATA ===")
print(f"CONTROL_NUM Used: {CONTROL_NUM}")
print(f"STUDENT_MAJOR Used: {STUDENT_MAJOR}")
print(f"Expanded Stream Length: {len(stream)}")
print(f"Total Applications Processed for TUP Manila: {len(campus_programs['TUP Manila'])}")
print(f"Top Requested Program at TUP Manila: {top_program}")
print(f"Exact Frequency of Top Program: {top_freq}")
```

```python
#APPENDIX VARIABLES

SEED_NUM = 0
STUDENT_MAJOR = "BSECE"
FAVORITE_ARTIST = "BINI"
LOCAL_HAZARD = "TYPHOON"
CONTROL_NUM = max(3, SEED_NUM)

print("Variables initialized successfully.")
```
