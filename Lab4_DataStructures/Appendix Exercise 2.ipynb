```python
#EXERCISE 2: Jaccard Similarity Engine

#Official Festival Lineup
festival = {"Ben&Ben", "SB19", "Bini", "Eraserheads", "REDVELVET", "Zild", f"Indie Artist {CONTROL_NUM}"}

#User Profiles
user_a = {"Ben&Ben", "Bini", "Maki", "Dionela", "REDVELVET"}
user_b = {"SB19", "Eraserheads", "Zild", f"Indie Artist {CONTROL_NUM}", "Parokya ni Edgar"}

#Data Expansion: add CONTROL_NUM new artists (CONTROL_NUM = 3)
festival.update({"ArtistX", "ArtistY", "ArtistZ"})

#Jaccard Similarity Function
def jaccard_similarity(set1, set2):
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    similarity = (intersection / union) * 100
    return intersection, union, similarity

#Calculate similarities
a_intersection, a_union, a_similarity = jaccard_similarity(user_a, festival)
b_intersection, b_union, b_similarity = jaccard_similarity(user_b, festival)

#Dealbreakers for User A
dealbreakers_a = user_a.difference(festival)

#PRINT ASSESSMENT DATA

print("=== ASSESSMENT DATA ===")
print(f"Total Artists in Expanded Festival: {len(festival)}")
print(f"Length of Intersection (User A & Festival): {a_intersection}")
print(f"Length of Union (User A & Festival): {a_union}")
print(f"User A Jaccard Similarity (%): {a_similarity:.2f}")
print(f"User B Jaccard Similarity (%): {b_similarity:.2f}")
print(f"User A Dealbreaker Artists: {dealbreakers_a}")
```
