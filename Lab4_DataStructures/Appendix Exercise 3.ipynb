```python
#EXERCISE 3: Sliding Window Filter

from collections import deque

#Hazard stream (simulated reports)
hazard_stream = [
    "FLOOD", "TYPHOON", "EARTHQUAKE", "LANDSLIDE", "FIRE",
    LOCAL_HAZARD, "VOLCANO", "TYPHOON", "FLOOD", "EARTHQUAKE"
]

#Window size based on CONTROL_NUM
window_size = CONTROL_NUM

#Sliding window filter
window = deque(maxlen=window_size)
timestamps = []   # simulate timestamps as sequential integers
filtered_reports = []

for i, hazard in enumerate(hazard_stream, start=1):
    window.append(hazard)
    timestamps.append(i)  # simulate timestamp
    filtered_reports.append(set(window))

#PRINT ASSESSMENT DATA
print("=== ASSESSMENT DATA ===")
print(f"LOCAL_HAZARD Used: {LOCAL_HAZARD}")
print(f"Total Burst Size (After adding custom alerts): {len(hazard_stream)}")

#Check if WEATHER at Timestamp CONTROL_NUM+3 appended
target_timestamp = CONTROL_NUM + 3
did_weather = "WEATHER" in hazard_stream[:target_timestamp]
print(f'Did "WEATHER" at Timestamp {CONTROL_NUM + 3} append? (Y/N): {"Y" if did_weather else "N"}')

#Check if LOCAL_HAZARD appended
did_local_hazard = LOCAL_HAZARD in hazard_stream
print(f"Did LOCAL_HAZARD append? (Y/N): {'Y' if did_local_hazard else 'N'}")

print(f"Final Deque Length: {len(window)}")
print(f"Final Sum of Timestamps in Deque: {sum(timestamps[-window_size:])}")
print(f"Exact Final Deque Output: {list(window)}")
```
